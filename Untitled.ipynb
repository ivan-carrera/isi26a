## Paso 1 — Construir Conocimiento desde los Datos

Los Sistemas de Procesamiento de Transacciones (TPS) capturan datos atómicos
en tiempo real. Los Sistemas de Información Gerencial (MIS) toman esos datos y
los transforman en información estructurada para apoyar decisiones. El ciclo
completo sigue la jerarquía: **Data Items → Information → Knowledge**.

---

**Los 3 KPIs del proyecto y su matriz conceptual:**

---

### KPI 1 — Tendencia de Ingresos Mensuales

| Nivel | Descripción |
|---|---|
| **Data Items** | Campos `total` (valor numérico crudo de cada venta) y `fecha` (cadena de texto `YYYY-MM-DD`) almacenados fila por fila en `transacciones.csv`. Cada registro representa un evento atómico aislado, sin ningún contexto temporal agregado. |
| **Information** | Al extraer el año y mes del campo `fecha` y agrupar todas las transacciones por ese período con `groupby`, se suman todos los valores de `total` por mes. El resultado es una serie temporal de hasta 12 puntos que muestra cuánto dinero generó el negocio en cada mes calendario. |
| **Knowledge** | Si la curva presenta una caída sostenida durante 2 o más meses consecutivos, gerencia debe revisar la estrategia de precios o la rotación de inventario. Si un mes supera el promedio histórico en más del 30%, se identifica como temporada alta y se puede planificar reabastecimiento preventivo con anticipación. |

---

### KPI 2 — Top 10 Productos Más Vendidos (por unidades despachadas)

| Nivel | Descripción |
|---|---|
| **Data Items** | Campos `codigo_producto`, `nombre_producto` y `cantidad` registrados línea por línea en `detalles_venta.csv`. Cada fila representa una línea de detalle atómica: qué producto específico, cuántas unidades, dentro de qué transacción. |
| **Information** | Al agrupar por `codigo_producto` y sumar el campo `cantidad` a través de todas las transacciones del período, se obtiene el volumen total de unidades despachadas por SKU. Ordenar ese resultado de mayor a menor y tomar los primeros 10 produce el ranking de productos estrella del negocio. Un cruce adicional con `inventario.csv` permite identificar cuáles de esos productos líderes tienen stock crítico. |
| **Knowledge** | Si un producto del top 10 tiene stock menor a 10 unidades (alerta visible en rojo en el gráfico), gerencia debe emitir una orden de reabastecimiento de emergencia antes de que se agote. Si un producto nuevo escala posiciones en el ranking durante períodos consecutivos, es candidato a recibir mejor ubicación en estantería y negociación de mayor volumen con el proveedor. |

---

### KPI 3 — Distribución de Ventas por Método de Pago

| Nivel | Descripción |
|---|---|
| **Data Items** | Campo `metodo_pago` en `transacciones.csv` (valores categóricos posibles: `Efectivo`, `Tarjeta`, `Transferencia`, `QR`) y el campo `total` monetario asociado a cada una de esas transacciones. |
| **Information** | Agrupando por `metodo_pago` se calcula tanto el conteo de transacciones por canal como el monto total acumulado en cada uno. El porcentaje de cada método sobre el total de transacciones revela los hábitos de pago predominantes de la clientela del minimarket. |
| **Knowledge** | Si el efectivo supera el 55% de las transacciones de forma sostenida, gerencia evalúa instalar terminales POS o QR adicionales y puede diseñar un pequeño incentivo (descuento del 2%) para clientes que paguen con medios digitales. Si el canal QR crece más de 5 puntos porcentuales en un trimestre, se justifica renegociar las comisiones con el proveedor de pagos digitales aprovechando el mayor volumen como poder de negociación.  |

---



## Paso 2 — Mapeo de Atributos y Boceto de Interfaz

### 2.1 Mapeo en el Esquema de Base de Datos

Las tablas del TPS que alimentan este MIS son las siguientes:

| Tabla CSV | Entidad lógica | Atributos utilizados | Rol en el MIS |
|---|---|---|---|
| `transacciones.csv` | Cabecera de venta | `id_transaccion`, `fecha`, `total`, `metodo_pago`, `nombre_cliente` | Base de KPI 1 (tendencia) y KPI 3 (métodos de pago) |
| `detalles_venta.csv` | Línea de detalle de venta | `id_transaccion`, `codigo_producto`, `nombre_producto`, `cantidad`, `subtotal_linea` | Base de KPI 2 (productos más vendidos) |
| `inventario.csv` | Stock actual por producto | `codigo`, `stock` | Alerta cruzada en KPI 2 para detectar productos líderes con stock crítico |
| `usuarios.csv` | Clientes registrados | `cedula` (solo conteo) | Métrica de tarjeta resumen: total de clientes activos |
| `productos.csv` | Catálogo de productos | `codigo`, `nombre`, `categoria` | Enriquecimiento de nombres en KPI 2 si fuera necesario |

**Relaciones entre entidades:**

- `transacciones.csv` y `detalles_venta.csv` se relacionan mediante `id_transaccion` en una relación **1 a N** (una transacción tiene uno o más productos).
- `detalles_venta.csv` e `inventario.csv` se relacionan mediante `codigo_producto` = `codigo` en una relación **N a 1** (muchas líneas de detalle apuntan al mismo producto en inventario).

### 2.2 Boceto Visual y Justificación

El dashboard está organizado en tres zonas:

**Zona 1 — Fila de tarjetas resumen**  
Cuatro tarjetas numéricas de alto impacto muestran: total de transacciones,
ingresos totales del período, ticket promedio por factura, y clientes registrados.
Se eligen tarjetas numéricas porque permiten al gerente leer el estado global
del negocio de un solo vistazo, sin necesidad de interpretar un gráfico.

**Zona 2 — Sección central izquierda: Gráfico de líneas (KPI 1)**  
Se elige el gráfico de líneas para representar la tendencia de ingresos porque
este tipo de visualización comunica dirección (crecimiento o caída),
velocidad del cambio (pendiente), y permite identificar estacionalidades.
El área sombreada bajo la línea refuerza visualmente la magnitud acumulada.

**Zona 3 — Sección central derecha: Gráfico de torta (KPI 3)**  
El gráfico de torta es adecuado cuando se distribuye un todo entre pocas
categorías mutuamente excluyentes (en este caso 4 métodos de pago).
Los porcentajes impresos directamente en cada segmento eliminan la necesidad
de leer una tabla separada.

**Zona 4 — Fila inferior: Barras horizontales (KPI 2)**  
Las barras horizontales son superiores a las verticales cuando los nombres
de las categorías son largos (como nombres de productos). El color rojo
marca automáticamente los productos con stock crítico, convirtiendo el gráfico
en una herramienta de alerta operativa además de un ranking.

## Paso 3 — Implementación del MIS en Python

El siguiente bloque de código constituye la implementación técnica del MIS:



```python
# ══════════════════════════════════════════════════════════════════
#  MIS QuickShop — Dashboard de KPIs
# ══════════════════════════════════════════════════════════════════

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib.gridspec import GridSpec
from matplotlib.patches import Patch
import os

# ── Rutas a los CSV del TPS ──────────────────────────────────────
RUTA_TRANSACCIONES = "transacciones.csv"
RUTA_DETALLES      = "detalles_venta.csv"
RUTA_INVENTARIO    = "inventario.csv"
RUTA_USUARIOS      = "usuarios.csv"
RUTA_PRODUCTOS     = "productos.csv"

# ── Verificación de archivos antes de continuar ──────────────────
for archivo in [RUTA_TRANSACCIONES, RUTA_DETALLES,
                RUTA_INVENTARIO, RUTA_USUARIOS, RUTA_PRODUCTOS]:
    if not os.path.exists(archivo):
        raise FileNotFoundError(
            f"\n Archivo '{archivo}' no encontrado.\n"
            "Asegúrate de haber ejecutado el TPS (main.py) primero\n"
            "y de que este notebook esté en la misma carpeta que los CSV."
        )

# ── Carga de datos ───────────────────────────────────────────────
df_tx  = pd.read_csv(RUTA_TRANSACCIONES, parse_dates=["fecha"])
df_det = pd.read_csv(RUTA_DETALLES)
df_inv = pd.read_csv(RUTA_INVENTARIO)
df_usr = pd.read_csv(RUTA_USUARIOS)

print(f" Transacciones cargadas  : {len(df_tx):>5} registros")
print(f" Detalles de venta       : {len(df_det):>5} registros")
print(f" Productos en inventario : {len(df_inv):>5} registros")
print(f" Clientes registrados    : {len(df_usr):>5} registros")
print()

# ════════════════════════════════════════════════════════════════
#  KPI 1 — Tendencia de ingresos mensuales (últimos 12 meses)
# ════════════════════════════════════════════════════════════════
df_tx["anio_mes"] = df_tx["fecha"].dt.to_period("M")

kpi1 = (
    df_tx.groupby("anio_mes")["total"]
    .sum()
    .sort_index()
    .tail(12)
    .reset_index()
)
kpi1.columns = ["mes", "ingresos"]
kpi1["mes_str"] = kpi1["mes"].dt.strftime("%b %Y")

print(f"📊 KPI 1 — Meses disponibles: {len(kpi1)}")
print(kpi1[["mes_str", "ingresos"]].to_string(index=False))
print()

# ════════════════════════════════════════════════════════════════
#  KPI 2 — Top 10 productos más vendidos (por unidades)
# ════════════════════════════════════════════════════════════════
kpi2 = (
    df_det.groupby(["codigo_producto", "nombre_producto"])["cantidad"]
    .sum()
    .reset_index()
    .sort_values("cantidad", ascending=False)
    .head(10)
    .reset_index(drop=True)
)
kpi2.index = kpi2.index + 1

# Cruce con inventario para detectar stock bajo
kpi2 = kpi2.merge(
    df_inv.rename(columns={"codigo": "codigo_producto"}),
    on="codigo_producto",
    how="left"
)
kpi2["alerta"] = kpi2["stock"].fillna(0) < 10

print(f"📊 KPI 2 — Top 10 productos:")
print(kpi2[["nombre_producto", "cantidad", "stock", "alerta"]].to_string())
print()

# ════════════════════════════════════════════════════════════════
#  KPI 3 — Distribución de ventas por método de pago
# ════════════════════════════════════════════════════════════════
kpi3_conteo = (df_tx.groupby("metodo_pago")["id_transaccion"]
               .count().reset_index())
kpi3_monto  = (df_tx.groupby("metodo_pago")["total"]
               .sum().reset_index())
kpi3 = kpi3_conteo.merge(kpi3_monto, on="metodo_pago")
kpi3.columns = ["metodo", "num_ventas", "monto_total"]
kpi3 = kpi3.sort_values("num_ventas", ascending=False).reset_index(drop=True)

print(f"📊 KPI 3 — Métodos de pago:")
print(kpi3.to_string(index=False))
print()

# ── Métricas para tarjetas resumen ──────────────────────────────
total_tx       = len(df_tx)
ingresos_total = df_tx["total"].sum()
ticket_prom    = df_tx["total"].mean()
clientes_reg   = len(df_usr)

# ════════════════════════════════════════════════════════════════
#  PALETA DE COLORES DEL DASHBOARD
# ════════════════════════════════════════════════════════════════
COLORES = {
    "azul"    : "#2563EB",
    "verde"   : "#16A34A",
    "naranja" : "#D97706",
    "morado"  : "#7C3AED",
    "rojo"    : "#DC2626",
    "gris"    : "#6B7280",
    "fondo"   : "#F8FAFC",
    "panel"   : "#FFFFFF",
    "borde"   : "#E2E8F0",
}

# ════════════════════════════════════════════════════════════════
#  CONSTRUCCIÓN DEL DASHBOARD
# ════════════════════════════════════════════════════════════════
fig = plt.figure(figsize=(20, 12), facecolor=COLORES["fondo"])
fig.suptitle(
    "Dashboard MIS — QuickShop MiniMarket\n"
    "Reporte de Indicadores Clave de Rendimiento (KPIs) | Últimos 12 meses",
    fontsize=16, fontweight="bold", color="#1E293B", y=1.02
)

gs = GridSpec(
    nrows=3, ncols=3,
    figure=fig,
    height_ratios=[0.18, 1, 1],
    hspace=0.50,
    wspace=0.35,
    left=0.06, right=0.97, top=0.80, bottom=0.06
)

# ─────────────────────────────────────────────────────────────────
#  TARJETAS RESUMEN 
# ─────────────────────────────────────────────────────────────────
tarjetas = [
    ("Total transacciones",  f"{total_tx:,}",             COLORES["azul"]),
    ("Ingresos totales",     f"${ingresos_total:,.2f}",    COLORES["verde"]),
    ("Ticket promedio",      f"${ticket_prom:.2f}",        COLORES["naranja"]),
    ("Clientes registrados", f"{clientes_reg:,}",          COLORES["morado"]),
]

for i, (titulo, valor, color) in enumerate(tarjetas):
    x_pos = 0.06 + i * 0.235
    ax_card = fig.add_axes([x_pos, 0.84, 0.215, 0.075])
    ax_card.set_facecolor(COLORES["panel"])
    for spine in ax_card.spines.values():
        spine.set_edgecolor(COLORES["borde"])
        spine.set_linewidth(1.2)
    ax_card.set_xticks([])
    ax_card.set_yticks([])
    ax_card.text(
        0.5, 0.70, titulo,
        ha="center", va="center",
        fontsize=9, color=COLORES["gris"],
        transform=ax_card.transAxes
    )
    ax_card.text(
        0.5, 0.25, valor,
        ha="center", va="center",
        fontsize=18, fontweight="bold", color=color,
        transform=ax_card.transAxes
    )
    ax_card.axvline(x=0, color=color, linewidth=5)

# ─────────────────────────────────────────────────────────────────
#  KPI 1 — Gráfico de líneas
# ─────────────────────────────────────────────────────────────────
ax1 = fig.add_subplot(gs[1, :2])
ax1.set_facecolor(COLORES["panel"])

ax1.plot(
    kpi1["mes_str"], kpi1["ingresos"],
    color=COLORES["azul"],
    linewidth=2.5,
    marker="o",
    markersize=7,
    markerfacecolor="white",
    markeredgewidth=2,
    markeredgecolor=COLORES["azul"],
    label="Ingresos mensuales",
    zorder=3
)

ax1.fill_between(
    kpi1["mes_str"],
    kpi1["ingresos"],
    alpha=0.10,
    color=COLORES["azul"]
)

prom_kpi1 = kpi1["ingresos"].mean()
ax1.axhline(
    prom_kpi1,
    color=COLORES["naranja"],
    linewidth=1.4,
    linestyle="--",
    alpha=0.8,
    label=f"Promedio: ${prom_kpi1:,.2f}"
)

# Anotación del mes con mayor ingreso
idx_max = kpi1["ingresos"].idxmax()
ax1.annotate(
    f"Máximo\n${kpi1.loc[idx_max, 'ingresos']:,.2f}",
    xy=(kpi1.loc[idx_max, "mes_str"], kpi1.loc[idx_max, "ingresos"]),
    xytext=(0, 28),
    textcoords="offset points",
    ha="center",
    fontsize=8.5,
    color=COLORES["verde"],
    fontweight="bold",
    arrowprops=dict(
        arrowstyle="->",
        color=COLORES["verde"],
        lw=1.3
    )
)

# Anotación del mes con menor ingreso
idx_min = kpi1["ingresos"].idxmin()
ax1.annotate(
    f"Mínimo\n${kpi1.loc[idx_min, 'ingresos']:,.2f}",
    xy=(kpi1.loc[idx_min, "mes_str"], kpi1.loc[idx_min, "ingresos"]),
    xytext=(0, -36),
    textcoords="offset points",
    ha="center",
    fontsize=8.5,
    color=COLORES["rojo"],
    fontweight="bold",
    arrowprops=dict(
        arrowstyle="->",
        color=COLORES["rojo"],
        lw=1.3
    )
)

ax1.set_title(
    "KPI 1 — Tendencia de Ingresos Mensuales",
    fontsize=12, fontweight="bold", color="#1E293B", pad=12
)
ax1.set_xlabel("Mes", fontsize=9, color=COLORES["gris"])
ax1.set_ylabel("Ingresos ($)", fontsize=9, color=COLORES["gris"])
ax1.tick_params(axis="x", rotation=38, labelsize=8)
ax1.tick_params(axis="y", labelsize=8)
ax1.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f"${x:,.0f}")
)
ax1.grid(axis="y", linestyle="--", alpha=0.4, color=COLORES["borde"])
ax1.spines[["top", "right"]].set_visible(False)
ax1.spines[["left", "bottom"]].set_edgecolor(COLORES["borde"])
ax1.legend(fontsize=9, loc="upper left", framealpha=0.8)

# ─────────────────────────────────────────────────────────────────
#  KPI 3 — Gráfico de torta
# ─────────────────────────────────────────────────────────────────
ax3 = fig.add_subplot(gs[1, 2])
ax3.set_facecolor(COLORES["panel"])

colores_pie = [
    COLORES["azul"],
    COLORES["naranja"],
    COLORES["verde"],
    COLORES["morado"]
][:len(kpi3)]

wedges, texts, autotexts = ax3.pie(
    kpi3["num_ventas"],
    labels=None,
    autopct="%1.1f%%",
    startangle=90,
    colors=colores_pie,
    pctdistance=0.72,
    wedgeprops={"linewidth": 2, "edgecolor": "white"},
    explode=[0.04] * len(kpi3)
)
for at in autotexts:
    at.set_fontsize(8.5)
    at.set_color("white")
    at.set_fontweight("bold")

leyendas = [
    f"{r['metodo']}  —  {r['num_ventas']} ventas  |  ${r['monto_total']:,.2f}"
    for _, r in kpi3.iterrows()
]
ax3.legend(
    wedges, leyendas,
    loc="lower center",
    bbox_to_anchor=(0.5, -0.32),
    fontsize=7.5,
    framealpha=0.85,
    title="Método de pago",
    title_fontsize=8.5
)
ax3.set_title(
    "KPI 3 — Método de Pago",
    fontsize=12, fontweight="bold", color="#1E293B", pad=12
)

# ─────────────────────────────────────────────────────────────────
#  KPI 2 — Barras horizontales
# ─────────────────────────────────────────────────────────────────
ax2 = fig.add_subplot(gs[2, :])
ax2.set_facecolor(COLORES["panel"])

nombres_cortos = [
    n[:40] + "…" if len(n) > 40 else n
    for n in kpi2["nombre_producto"]
]
colores_barras = [
    COLORES["rojo"] if alerta else COLORES["verde"]
    for alerta in kpi2["alerta"]
]

bars = ax2.barh(
    nombres_cortos[::-1],
    kpi2["cantidad"][::-1],
    color=colores_barras[::-1],
    edgecolor="white",
    linewidth=0.8,
    height=0.65
)

max_val = kpi2["cantidad"].max()

for bar, val, alerta, stock in zip(
    bars,
    kpi2["cantidad"][::-1],
    kpi2["alerta"][::-1],
    kpi2["stock"].fillna(0).astype(int)[::-1]
):
    etiqueta = f"  {int(val)} unidades"
    if alerta:
        etiqueta += f"   ⚠ stock actual: {stock}"
    ax2.text(
        bar.get_width() + max_val * 0.005,
        bar.get_y() + bar.get_height() / 2,
        etiqueta,
        va="center",
        ha="left",
        fontsize=8.5,
        color=COLORES["rojo"] if alerta else COLORES["gris"]
    )

ax2.set_title(
    "KPI 2 — Top 10 Productos Más Vendidos (unidades despachadas en el período)",
    fontsize=12, fontweight="bold", color="#1E293B", pad=12
)
ax2.set_xlabel("Unidades vendidas (total del período analizado)",
               fontsize=9, color=COLORES["gris"])
ax2.tick_params(axis="y", labelsize=9)
ax2.tick_params(axis="x", labelsize=8)
ax2.grid(axis="x", linestyle="--", alpha=0.35, color=COLORES["borde"])
ax2.spines[["top", "right"]].set_visible(False)
ax2.spines[["left", "bottom"]].set_edgecolor(COLORES["borde"])
ax2.set_xlim(0, max_val * 1.28)

leyenda_elementos = [
    Patch(color=COLORES["verde"],
          label="Stock normal — sin alerta"),
    Patch(color=COLORES["rojo"],
          label="⚠ Stock bajo (< 10 unidades) — requiere reabastecimiento urgente"),
]
ax2.legend(
    handles=leyenda_elementos,
    loc="lower right",
    fontsize=8.5,
    framealpha=0.85
)

# ── Nota al pie del dashboard ────────────────────────────────────
fecha_min = df_tx["fecha"].min().strftime("%d/%m/%Y")
fecha_max = df_tx["fecha"].max().strftime("%d/%m/%Y")
fig.text(
    0.5, 0.005,
    f"MIS QuickShop v1.0  |  Fuente: transacciones.csv + detalles_venta.csv + "
    f"inventario.csv  |  Período: {fecha_min} → {fecha_max}  |  "
    f"Total registros analizados: {len(df_tx):,} transacciones / "
    f"{len(df_det):,} líneas de detalle",
    ha="center",
    fontsize=7.5,
    color=COLORES["gris"]
)

# ── Guardar y mostrar ────────────────────────────────────────────
plt.savefig(
    "dashboard_mis_quickshop.png",
    dpi=150,
    bbox_inches="tight",
    facecolor=COLORES["fondo"]
)
plt.show()
print("\n Dashboard generado y guardado como: dashboard_mis_quickshop.png")
```

## Paso 4 — Validación e Interpretación Gerencial

### 4.1 Matriz de Atributos de Valor

| KPI | Simplicidad | Oportunidad (Timeliness) | Accesibilidad |

Cada punto representa exactamente un mes. | La consulta procesa los CSV locales en menos de 1 segundo. El gerente puede ejecutarla al inicio de cada jornada y tener el panorama del mes anterior actualizado. | La pendiente de la curva comunica crecimiento o caída sin necesidad de leer valores numéricos. La alerta de stock cruzada con inventario permite actuar antes de que ocurra el desabastecimiento, no después. | El ranking visual 1-10 es completamente autoexplicativo: la barra más larga corresponde al producto más vendido, sin necesidad de formación técnica previa. |

Los porcentajes están impresos directamente sobre cada segmento, eliminando la necesidad de leer una tabla separada. | Agrupación ejecutada en milisegundos. Útil para decisiones operativas del mismo día (por ejemplo, detectar que el lector de tarjetas falló y las ventas cayeron a cero en ese canal). | El gráfico de torta es el formato más universalmente reconocible para distribuciones proporcionales. Cualquier persona del área administrativa lo interpreta sin capacitación. |

### 4.2 Documentación de Interpretación Gerencial

**KPI 1 — Tendencia de Ingresos Mensuales**

Si este indicador clave de rendimiento muestra un comportamiento negativo
(caída de ingresos por dos o más meses consecutivos), la acción correctiva
que la gerencia debe ejecutar de inmediato es revisar la política de precios,
identificar si hay productos de alta rotación con bajo margen que estén
erosionando la rentabilidad, y lanzar promociones focalizadas en las categorías
con mayor caída de ventas. Si el indicador muestra un pico estacional
(crecimiento superior al 25% respecto al mes anterior), la acción estratégica
es asegurar el nivel de inventario con anticipación para no perder ventas
por desabastecimiento durante ese período de alta demanda.

---

**KPI 2 — Top 10 Productos Más Vendidos**

Si este indicador clave de rendimiento muestra que uno o más productos del
top 10 tienen una alerta de stock bajo (barra en rojo, stock < 10 unidades),
la acción correctiva que la gerencia debe ejecutar de inmediato es emitir
una orden de reabastecimiento de emergencia a los proveedores correspondientes,
priorizando esos SKUs sobre el resto del catálogo. Si el indicador muestra
que un producto nuevo escala posiciones en el ranking durante varios períodos
consecutivos, la acción estratégica es aumentar su visibilidad en el punto de
venta (reubicación en estantería preferente) y negociar un volumen de compra
mayor con el proveedor para obtener mejor precio unitario.

**KPI 3 — Distribución por Método de Pago**

Si este indicador clave de rendimiento muestra que el efectivo supera el 55%
de las transacciones de forma sostenida, la acción correctiva que la gerencia
debe ejecutar de inmediato es evaluar la instalación de terminales POS o QR
adicionales y diseñar un incentivo para clientes que paguen con medios digitales
(descuento del 2-3%), reduciendo así el riesgo operativo del manejo físico
de dinero y los costos de cuadre de caja. Si el canal QR crece más de 5 puntos
porcentuales en un trimestre, la acción estratégica es renegociar las comisiones
con el proveedor de pagos digitales aprovechando el mayor volumen
como poder de negociación.
